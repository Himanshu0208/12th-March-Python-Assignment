# __Ques 2__
__Null Hypothesis H0__=The observed qquantty follws the excepted output <br>
__Alternate Hypothesis H1__=The observed quantity is entirely diffrent from the expected output

```python
import scipy.stats as stat

expected=[20,20,20,10,10,20]
observed=[25,25,10,10,20,10]
significance_value=0.05
dof=len(observed)-1

chi_square_test , p_value=stat.chisquare(f_obs=observed , f_exp=expected)

chi_critical_value=stat.chi2.ppf(1-significance_value,dof)

print(chi_square_test , chi_critical_value)

if chi_square_test>chi_critical_value:
    print('Reject The Null Hypothsis')
else :
    print('Fail to reject the null hypothesis')
```

# __Ques 3__
__Null Hypothesis H0__=Both a and b are same <br>
__Alternate Hypothesis H1__=Bothe a and are diffrenet

```python
from scipy.stats import chi2_contingency

# create the contingency table
observed = [[20, 15],
            [10, 25],
            [15, 20]]

significance=0.05

# calculate the chi-square statistic and p-value
chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("P-value:", p)

if p<significance:
    print('Reject The Null Hypothsis')
else :
    print('Fail to reject the null hypothesis')
```

# __Ques 6__

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# set the degrees of freedom
df = 10

# set the x range for the plot
x = np.linspace(0, 30, 500)

# create the chi-square distribution with 10 degrees of freedom
chisq = chi2(df)

# plot the chi-square distribution
plt.plot(x, chisq.pdf(x), label="Chi-square PDF")

# shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 100)
y_shade = chisq.pdf(x_shade)
plt.fill_between(x_shade, y_shade, 0, alpha=0.5)

# set the x and y labels and title
plt.xlabel("Chi-square statistic")
plt.ylabel("Probability density")
plt.title("Chi-square distribution with 10 degrees of freedom")

# add a legend to the plot
plt.legend()

# show the plot
plt.show()
```
